 DATA ANALYSE :Comparaison canopée greehill et MNHmax IGN sur Marseille

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Lire le fichier CSV
df = pd.read_csv(r'c:\Users\PC\pipeline-trees\data\in\Greehil_especes_Lidar_canopee_zonal_statistics_filtre2.csv', sep=';')
df.columns = df.columns.str.strip()  # Nettoyer les noms de colonnes

# 2. Conversion des colonnes en numériques
df['height'] = pd.to_numeric(df['height'], errors='coerce')      # Greehill
df['MNH_max'] = pd.to_numeric(df['MNH_max'], errors='coerce')    # IGN

# 3. Nettoyage : supprimer NaN et valeurs négatives
df_clean = df.dropna(subset=['height', 'MNH_max'])
df_clean = df_clean[(df_clean['height'] >= 0) & (df_clean['MNH_max'] >= 0)]

# 4. Définir les variables
x = df_clean['height']      # Hauteur mesurée Greehill
y = df_clean['MNH_max']     # Hauteur estimée IGN

# Formule analytique : a = sum(x*y) / sum(x^2)
a = np.sum(x * y) / np.sum(x ** 2)

# Calcul des valeurs ajustées
y_pred = a * x

# Coefficient de détermination R² pour régression par l'origine
ss_tot = np.sum(y ** 2)
ss_res = np.sum((y - y_pred) ** 2)
r_squared = 1 - ss_res / ss_tot


x_line = np.linspace(0, x.max(), 300)
y_line = a * x_line

# 7. Tracé du graphique
plt.figure(figsize=(14, 8))
plt.scatter(x, y, alpha=0.7, s=30, color='#13688c', label='Données')

# Droite de régression (pointillée noire)
plt.plot(x_line, y_line, color='black', linestyle='dotted', linewidth=2, label='Régression')

# Annotation de la régression
plt.text(x.max() * 0.1, y.max() * 0.9,
         f'y = {a:.4f}x\n$R^2$ = {r_squared:.4f}',
         fontsize=12)

# Titre et axes
plt.title(f'Comparison of canopies measured by Greehill and IGN LIDAR over Marseille({len(df_clean)} points)', fontsize=15)
plt.xlabel('Canopy height measured by Greenhill (m)', fontsize=13)
plt.ylabel('Canopy height (MNHmax) estimated by IGN LIDAR (m)', fontsize=13)
plt.grid(True)
plt.xlim(0, x.max() + 2)
plt.ylim(0, y.max() + 2)
plt.legend()
plt.tight_layout()
plt.show()
```
